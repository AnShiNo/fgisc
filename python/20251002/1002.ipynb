```python
#如果想跳過某次動作
#可以用 continue，但下一次計次仍會執行動作
#如果想跳出迴圈，則是用 break
for i in range(0,10,1):
	if i%2==0:
		continue
	print(i)
```

```python
for i in range(0,10,1):
    for j in range(0,10,1):
            print(0,end="")
    print("\n")
#巢狀迴圈: 迴圈裡面有迴圈，則會從最內圈開始執行。
#for迴圈是固定次數的迴圈
#range(起始值，結束值，間隔)
#所以上面會輸出0~9
```

```python
for i in range(1,10):
	for j in range(1,10):
		print(i*j,end="\t")
	print("\n")
```
