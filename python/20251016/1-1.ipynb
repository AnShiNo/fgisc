```python
#List類似於其他程式語言的 array (陣列)
#可以儲存多筆資料(顧名思義對吧他是清單)
#用 [ ] 將資料包起來
#使用索引值 (index) 編號存取，從 0 開始
#不同資料型態可以放在同個 List中

children=[1,2,1.5,"yeah"]
print(children[3]) #yeah
print(children[0]) #1
```

```python
#一維陣列
scores=[85,90,78,92,88]
print(scores[0])  # 85

#多維陣列
matrix=[
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
#matrix=[[1, 2, 3],[4, 5, 6],[7, 8, 9]]
print(matrix[0][0])  # 第一列第一行 → 1
print(matrix[2][1])  # 第三列第二行 → 8
```

```python
#在python中，可使用負數的索引值
#例如倒數第一個元素為 -1

children=[1,2,3,4,5,6,7,8,9,0]
print(children[-2]) #9
```

```python
#可以用 len() 取得串列長度
#如果是多維串列則會取最外層的長度

children=[1,2,3,4,5,6,7,8,9,0]
print(len(children)) #10
matrix=[[1,2,3],[4,5,6],[7,8,9]]
print(len(matrix)) #3
```

```python
#使用 slicing 可以從串列中取出一部份連續的資料
#有點類似for迴圈的概念
#list[起始位置索引值:結束位置索引值+1:間隔]

children=[1,2.5,"hello",4,0,60,0.1]
print(children[2:6:1]) #['hello',4,0,60] #2~5.間隔1
print(children[:3]) #[1,2.5,'hello'] #0~2間隔默認1
print(children[3:]) #[4,0,60,0.1] #3~6間隔默認1
print(children[:]) #[1,2.5,'hello',4,0,60,0.1] #全部
print(children[::2]) #[1,'hello',0,0.1] #全部 間隔2取
print(children[1::2]) #[2.5,4,60] #從1開始 間隔2取
print(children[::-1]) #反轉
```

```python
#list[起始位置索引值:結束位置索引值+1:間隔]
#也可以使用 slice 來修改值

children=[1,2.5,"hello",4,0,60,0.1]
children[1:3]=["a","cat"] #[1,"a","cat",4,0,60,0.1]
```

```python
#list[index]="要改的值"
#可以改值，但索引值不能超過串列範圍
children=[1,2,3,4]
children[2]="ok" #[1,2,"ok",4]
```

```python
#append ( )：加到串列尾端
#insert （index, value）：插入指定位置
#extend ( )：串接兩個列表
children=[1,2,3,4]
children.append('a') #[1,2,3,4,'a']
children.insert(2,5.5) #[1,2,5.5,3,4,'a']
adult=[9,8,7,6]
children.extend(adult)
print(children) #[1,2,5.5,3,4,'a',9,8,7,6]
#寫前面的列表.extend(後面的列表)
```

```python
#remove( )：刪除特定值（只刪相同元素的第一個）
#pop( )：刪除特定位置（無指定index則刪最後一個）
#del( )：切片刪除
#clear( )：清空串列
children=[1,2,2,3,4,5.5,60,700]
children.remove(2) #刪除第一個2 [1,2,3,4,5.5,60,700]
children.pop(6) #[1,2,3,4,5.5,60]
#為甚麼這樣可以輸出? 因為pop會回傳被刪除的值
```

```python
#使用 sort( ) 可以將串列從小到大排序
#若元素資料型態為string
#則會依照第一個字元的ASCII排序
#第一個相同則比較第二個
#不同資料型態不能比較

children=[40,55,76,80,31]
children.sort()
print(children) #[31,40,55,76,80]
```

```python
children=[1,2,3,4,5]
children.reverse()
print(children) #[5,4,3,2,1]

#使用 reverse( ) 可以將串列反轉
#要讓串列由大到小排列:
children.sort(reverse=True) 
#為甚麼要多這一行? 為甚麼不直接用reverse? 
#因為sort預設是從小到大排列 然後 再反轉
```

```python
children=[5,2,6,1,3,4]
children.reverse()
children.sort(reverse=True) 
print(children) 

#使用 reverse( ) 可以將串列反轉
#要讓串列由大到小排列在程式裡加這一行:
children.sort(reverse=True) 
#為甚麼要多這一行? 為甚麼不直接用reverse? 
#因為sort預設是從小到大排列 然後 再反轉
```

```python
#使用 index( ) 可以回傳一個元素的索引值
#甚麼是索引直? 就是位置編號
children=[1,2,3,4,5]
print(children.index(4)) #3
```

```python
#使用count( ) 回傳一個元素在串列中出現的次數

children=[1,2,3,3,5]
print(children.count(3)) #2
print(children.count(4)) #0
```

```python
#in : 回傳該元素是否在串列中
#not in : 回傳該元素是否不在串列中

children=[1,2,3,3,5]
print(1 in children) #True
print(4 in children) #False
print(4 not in children) #True
```

```python
#max : 回傳串列最大值
#min : 回傳串列最小值
#只有在串列中所有元素資料型態相同才適用
children=[1,2,3,3,5]
print(max(children)) #5
print(min(children)) #1
```

```python
#sum( ) 可以把串列中所有元素加起來
#但僅限於整數和浮點數
children=[1,2,3,3,5]
print(children.count(3)) #2
print(children.count(4)) #0
```
